# Simple Linear Regression

### Goal
Understand how simple linear regression works using Scikit-learn.

### Dataset
A small dataset containing canada's years and per capital income

### Steps
1. Import libraries

3. Train Linear Regression model
4. Predict values
5. Visualize results

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#load dataset
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
```

```python
#Model training
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
#Predict values
prediction = reg.predict([[2020]])
print("Predicted Per Capita Income in 2020:", prediction[0])
```
